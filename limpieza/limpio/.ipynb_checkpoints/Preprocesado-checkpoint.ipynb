```python
datos_familia=pd.read_csv('./bird_taxonomy_families.csv')

datos_espana=pd.read_csv('./controles_espana_merged.csv')

datos_extranjero=pd.read_csv('./controles_extranjeros_merged.csv')

datos_anillamiento = pd.read_csv('./anillamiento.csv')

datos_sin_anillamiento = pd.read_csv('./controles_sin_datos_de_anillamiento.csv')

datos_familia[:10]

```

```python
especies = datos_familia.SCI_NAME.to_list()
outliers_extranjeros = list()
outliers_espana = list()
outliers_anillamiento = list()
outliers_sin_anillamiento = list()

for i in datos_extranjero['ESPECIE']:
    if especies.count(i) == 0:
        outliers_extranjeros.append(i)
outliers_extranjeros = pd.Series(outliers_extranjeros).unique()

for i in datos_espana['ESPECIE']:
    if especies.count(i) == 0:
        outliers_espana.append(i)
outliers_espana = pd.Series(outliers_espana).unique()

for i in datos_anillamiento['NombreEspecie']:
    if especies.count(i) == 0:
        outliers_anillamiento.append(i)
outliers_anillamiento = pd.Series(outliers_anillamiento).unique()

for i in datos_sin_anillamiento['NombreEspecie']:
    if especies.count(i) == 0:
        outliers_sin_anillamiento.append(i)
outliers_sin_anillamiento = pd.Series(outliers_sin_anillamiento).unique()
```
